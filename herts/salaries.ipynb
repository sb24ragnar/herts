```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = 'data_science_salaries.csv'
df = pd.read_csv(file_path)
```

```python
USD_TO_GBP_RATE = 0.82
df['salary_in_gbp'] = df['salary_in_usd'] * USD_TO_GBP_RATE
```

```python
def plot_histogram(data):
    plt.figure(figsize=(10, 6))
    sns.histplot(data['salary_in_gbp'], bins=30, color='skyblue')
    plt.title('Distribution of Salary in GBP')
    plt.xlabel('Salary in GBP')
    plt.ylabel('Frequency')
    plt.savefig("histogram_gbp.png")
    plt.show()
    plt.close()

plot_histogram(df)
```

```python
def plot_scatter(data):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='work_year', y='salary_in_gbp', data=data, hue='experience_level', palette='viridis')
    plt.title('Yearly Salary in GBP by Experience Level')
    plt.xlabel('Work Year')
    plt.ylabel('Salary in GBP')
    plt.legend(title='Experience Level')
    plt.savefig("scatter_gbp.png")
    plt.show()
    plt.close()

plot_scatter(df)
```

```python
def plot_heatmap(data):
    plt.figure(figsize=(8, 6))
    correlation_matrix = data[['work_year', 'salary', 'salary_in_gbp']].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Correlation Heatmap')
    plt.savefig("heatmap_gbp.png")
    plt.show()
    plt.close()

plot_heatmap(df)
```
